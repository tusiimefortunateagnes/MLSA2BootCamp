```python
column_name = 'gender'

# Create a histogram
plt.hist(df[column_name], bins=10)  
plt.xlabel(column_name)
plt.ylabel("Frequency")
plt.title("Histogram of " + column_name)

# Display the histogram
plt.show()
```

```python
column_name = 'gender'

# Group the data by the values in the selected column
data_counts = df[column_name].value_counts()

# Create a pie chart
plt.pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title("Pie Chart of " + column_name)

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
plt.show()
```
